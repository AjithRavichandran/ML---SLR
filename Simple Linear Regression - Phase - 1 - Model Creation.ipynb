```python
import pandas as pd                #Pandas is a Python library used for working with data sets.
import matplotlib.pyplot as plt    #Matplotlib is a Python library that allows users to create visualizations like histograms, bar charts, scatter plots
```

```python
dataset=pd.read_csv('Salary_Data.csv')    #called the CSV file using pandas and kept the variable name as 'dataset'
```

```python
dataset 
```

```python
independent=dataset[['YearsExperience']]    #assigning the YearsExperience to independent variable
independent
```

```python
dependent=dataset[['Salary']]   #assigning the Salary to dependent variable
dependent
```

```python
plt.figure(figsize=(5,3))                        #giving size to the chart
plt.scatter(independent,dependent)               #plotting the variables
plt.xlabel('Years of Experience', fontsize=20)   #Giving font size to x
plt.ylabel('Salary', fontsize=20)                #Giving font size yo y
plt.show()
```

```python
from sklearn.model_selection import train_test_split   #from sklearn.model_selection we are going to call train_test_split to them into 4 variables as input train,input test,output train,output test
```

```python
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)  #for dividing into 4 variables we are going to give 4 parameters 1.independent,2.depedndent,3. test size (how much percent we are going to assign for the test)  
```

```python
from sklearn.linear_model import LinearRegression    # in this LinearRegression we are going to find the model
regressor=LinearRegression()                         # here LinearRegression class things assigned to regressor 
regressor.fit(input_train,output_train)              # here we are giving the input train and output train to find the model and we are assigning to regressor 
```

```python
weight=regressor.coef_              # checking the weight
bias=regressor.intercept_           # checking the bias
print(f'weight is {weight}')
print(f'bias is {bias}')
```

```python
pre=regressor.predict(input_test)   # going to predict by giving input test
```

```python
from sklearn.metrics import r2_score    #going to find the r2 score by giving the actual and predicted value
r2=r2_score(output_test,pre)
```

```python
r2
```

```python
import pickle                                                    #pickle is helps to store the model
```

```python
filename='Finale Model Simple Linear.sav'                        #assigning the filename
```

```python
pickle.dump(regressor,open(filename,'wb'))                       #assiging the model(regressor) to the pickle's file name
```

```python
load=pickle.load(open('Finale Model Simple Linear.sav','rb'))    #loading the data to the model which is in pickle
```

```python
result=load.predict([[2]])                                       #predicting the value and assigning it to the 'result' variable
```
